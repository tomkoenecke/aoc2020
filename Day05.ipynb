```python
# convert to binary
new = []
for element in data:
    newline =[]
    for letter in element:
        if letter == 'B' or letter == 'R':
            newline.append(1)
        else:
            newline.append(0)
    new.append(newline)
```

```python
def bin_to_int(bitstring):
    power = len(bitstring) - 1
    out = 0
    while len(bitstring) != 0:
        out += 2**power * bitstring.pop(0)
        power -= 1
    return out
```

# Part 1

```python
maxi = 0
ids = []
for element in new:
    row_bin = element[:7]
    row = bin_to_int(row_bin)
    col_bin = element[-3:]
    col = bin_to_int(col_bin)
    ids.append(row*8 + col)
    maxi = max(maxi, row*8 + col)
print(maxi)
```
