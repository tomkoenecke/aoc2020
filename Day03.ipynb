# Part 1

```python
def traverser(data, right, down):
    row = 0
    col = 0
    tree_count = 0
    height = len(data)
    width = len(data[0])
    while row < height:
        if col >= width:
            col -= width
        if data[row][col] == '#':
            tree_count += 1
        row += down
        col += right

    return tree_count
```

```python
print(traverser(data, 3, 1))
```
