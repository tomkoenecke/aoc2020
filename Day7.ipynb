# Part 2
observation: every bag type is described exactly once in the data!

```python
rules = dict()
for rule in data:
    parent, children = rule.split(' bags contain ')
    temp = children.split(', ')
    children = []
    for child in temp:
        child = child.rsplit(' ', 1)[0]
        children.append(child)
    rules[parent] = children
```

```python
def get_child_number(name):
    children = rules[name]
            
    # recursive exit
    if children[0] == 'no other':
        return 1
    
    res = 0
    for child in children:
        factor1 = int(child[0])
        factor2 = get_child_number(child[2:])
        res += factor1 * factor2
    return res+1 # +1 for parent bag

print(get_child_number('shiny gold')-1) # -1 for root bag 'shiny gold'
```
