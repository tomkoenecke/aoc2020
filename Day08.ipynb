```python
def is_terminal(data):
    i = 0
    accumulator = 0
    for _ in range(len(data)):
        if i == len(data):
            return True, accumulator
        
        cmd = data[i][:3]
        sign = data[i][4]
        steps = int(data[i][5:])
        if sign == '-':
            steps = -steps
        if cmd == 'nop':
            i +=1
        elif cmd == 'acc':
            i += 1
            accumulator += steps
        elif cmd == 'jmp':
            i += steps
    return False, 0
```
